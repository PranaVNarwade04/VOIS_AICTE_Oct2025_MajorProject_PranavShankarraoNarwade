## **1. Introduction**
Netflix is one of the world's most prominent global streaming platforms, with a massive library of movies and TV shows. As competition in the streaming industry (from Disney+, Amazon Prime, etc.) grows, it is crucial for Netflix to make data-driven decisions about its content strategy.

This project will analyze the Netflix dataset, which contains 7,789 records of movies and shows available on the platform, spanning the years 2008 to 2021. By analyzing this data, we aim to uncover trends in Netflix's content catalog that can inform strategic recommendations.

## **2. Problem Statement**
The specific problem for this analysis is: "Content Trends Analysis for Strategic Recommendations".

The aim is to uncover how Netflix’s content distribution—specifically (1) Movies vs. TV Shows, (2) genres, and (3) country contributions—has evolved over the years. This will help identify key audience preferences and provide strategic insights for global content expansion.

## **3. Project Objectives**

Based on the project's problem statement, our analysis will focus on the following three objectives:

1.   Analyze the distribution of **Movies vs. TV Shows** over the years.
2.   Identify the most **common genres** and how their popularity has
changed.
3.   Compare **country-wise** contributions to Netflix’s catalog.




## **4. Dataset Overview**
We will be using the Netflix Dataset.csv file. The key columns for our analysis will be

* **Category**: To distinguish between Movies and TV Shows.
* **Release Date**: To analyze trends over time.
* **Type**: To analyze genre distributions.

#### **Handling Duplicate Values**

```python
# Checking Duplicates in Dataset
df.duplicated().sum()
```

```python
initial_rows = df.shape[0]
print(f"Initial number of rows: {initial_rows}")
```

```python
# Drop rows that are complete duplicates
df.drop_duplicates(inplace=True)
print(f"Rows after dropping full duplicates: {df.shape[0]}")
```

```python
# Drop rows with duplicate titles, keeping the first instance
df.drop_duplicates(subset=['Title'], keep='first', inplace=True)
final_rows = df.shape[0]
print(f"Rows after dropping duplicate titles: {final_rows}")
print(f"Total rows removed: {initial_rows - final_rows}")
```

#### **Handling Null Values**

```python
df.isnull().sum()
```

```python
# Drop the few rows with missing essential data- Less Null Values present in these columns
df.dropna(subset=['Release_Date', 'Rating'], inplace=True)
print("Dropped rows with missing 'Release_Date' or 'Rating'.")
```

```python
# Fill 'Director', 'Cast', and 'Country' with 'Unknown'as we cannot fill wrong information nor we can remove as there are too many null values present in these columns
df['Director'].fillna("Unknown", inplace=True)
df['Cast'].fillna("Unknown", inplace=True)
df['Country'].fillna("Unknown", inplace=True)
print("Filled 'Director', 'Cast', and 'Country' with 'Unknown'.")
```

```python
# Checking If Null Values in Director, Cast and Country Columns are replaced with Unknown or Not ?
df.head()
```

```python
# Checking If Null values are replaced or Not
df.isnull().sum()
```

#### **Fixing Data Types and Creating New Columns**

```python
# The head output shows a leading space, e.g., " August 14, 2020"
# We must strip spaces AND provide the exact format
try:
    df['Release_Date'] = pd.to_datetime(df['Release_Date'].str.strip(), format='%B %d, %Y')
    print("Converted 'Release_Date' to datetime.")

    # Create new columns for analysis (required for Objective 1)
    df['Year_Added'] = df['Release_Date'].dt.year
    df['Month_Added'] = df['Release_Date'].dt.month_name()
    print("Created 'Year_Added' and 'Month_Added' columns.")
except Exception as e:
    print(f"Error converting date: {e}")
    print("There might be other date formats. Let's try to infer the format.")
    # Fallback if the format is mixed
    df['Release_Date'] = pd.to_datetime(df['Release_Date'].str.strip(), errors='coerce')
    # Drop any rows that failed conversion
    df.dropna(subset=['Release_Date'], inplace=True)
    # Re-create new columns
    df['Year_Added'] = df['Release_Date'].dt.year
    df['Month_Added'] = df['Release_Date'].dt.month_name()


# --- 4. Final Check ---
print("\n--- Final Cleaned Data Info ---")
df.info()
```

### **Analysis And Visualization**

#### **Objective 1**: Analyze Movies vs. TV Shows Over Time

```python
# Group by Year_Added and Category, then count the titles
content_over_time = df.groupby('Year_Added')['Category'].value_counts().unstack().fillna(0)
content_over_time = content_over_time.astype(int)

print("\n--- Content Added Per Year (Movies vs. TV Shows) ---")
# Sort by year to see the most recent data
print(content_over_time.sort_index(ascending=False).head())
```

```python
# --- Visualization ---
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Plot the line chart
content_over_time.plot(kind='line', marker='o', figsize=(12, 6))

plt.title('Content Added to Netflix Over Time (Movies vs. TV Shows)', fontsize=16)
plt.xlabel('Year Added', fontsize=12)
plt.ylabel('Number of Titles Added', fontsize=12)
plt.legend(title='Category', fontsize=10)
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Save the plot
plt.savefig('content_over_time.png')

print("\nVisualization saved as 'content_over_time.png'")
```

**Outcome 1**: The Line Graph shows massive growth in adding both movies and tv series arround in 2015-2016. The small decline took place at 2020. The more no TV series has been added recently (from 2016).

#### **Objective 2**: Identify the most common genres and how their popularity has changed.

**Top 10 Genre's**

```python
# As we have combined Genre's in type column we have to seperate them using comma delimeter
# Get the list of the top 10 genre names
from collections import Counter

print("--- Objective 2: Analyzing Top Genres ---")

# The 'Type' column has strings like "Dramas, International Movies"
# We need to split them and count each genre

# Use collections.Counter for an efficient way to count
genre_counter = Counter()

# Loop through each row's 'Type' string
for genres in df['Type']:
    # Split the string by comma and space, then update the counter
    genre_list = genres.split(', ')
    genre_counter.update(genre_list)
```

```python
# Get the 10 most common genres
top_10_genres = genre_counter.most_common(10)

# Convert to a DataFrame for easy plotting
df_top_10_genres = pd.DataFrame(top_10_genres, columns=['Genre', 'Count'])

print("\n--- Top 10 Most Common Genres on Netflix ---")
print(df_top_10_genres)
```

```python
# --- Visualization ---
plt.figure(figsize=(12, 8))

# Use seaborn for a horizontal bar plot
# We use y='Genre' and x='Count' for a horizontal chart
sns.barplot(x='Count', y='Genre', data=df_top_10_genres, palette='mako')

plt.title('Top 10 Most Common Genres on Netflix', fontsize=16)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.tight_layout()

# Save the plot
plt.savefig('top_10_genres.png')

print("\nVisualization saved as 'top_10_genres.png'")
```

**Outcome 2.1**- Netflix's growth in both in International and Drama's and Comedy Genre

```python
# Get the list of the top 10 genre names
top_10_genre_names = df_top_10_genres['Genre'].tolist()

# Create a new list of dictioprint("--- Objective 2 (Part 2): Analyzing Genre Trends Over Time ---")

# Get the list of the top 10 genre names
top_10_genre_names = df_top_10_genres['Genre'].tolist()

# Create a new list of dictionaries to build a new DataFrame
# This will "explode" the genres, but only for the ones we care about
genre_year_list = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    year = row['Year_Added']
    genres = row['Type'].split(', ')

    # Check if any of the genres in this row are in our top 10 list
    for g in genres:
        if g in top_10_genre_names:
            # If it is, add it to our new list with its year
            genre_year_list.append({'Year_Added': year, 'Genre': g})

# Convert the list to a new DataFrame
df_genre_trends = pd.DataFrame(genre_year_list)

# Now, we pivot the data: count Genres grouped by Year
genre_trends_pivot = df_genre_trends.groupby('Year_Added')['Genre'].value_counts().unstack().fillna(0)
genre_trends_pivot = genre_trends_pivot.astype(int)

print("\n--- Trend Data for Top 10 Genres (Last 5 Years) ---")
print(genre_trends_pivot.sort_index(ascending=False).head())


```

```python
# --- Visualization ---
plt.figure(figsize=(15, 10)) # A larger figure is needed

# Plot all 10 genres
genre_trends_pivot.plot(kind='line', marker='.', figsize=(15, 8))

plt.title('Popularity Trends of Top 10 Genres', fontsize=16)
plt.xlabel('Year Added', fontsize=12)
plt.ylabel('Number of Titles Added', fontsize=12)
plt.grid(True)

# Place the legend outside of the plot to avoid covering data
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout() # Adjust layout

# Save the plot
plt.savefig('top_10_genres_over_time.png')

print("\nVisualization saved as 'top_10_genres_over_time.png'")
# This will "explode" the genres, but only for the ones we care about
genre_year_list = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    year = row['Year_Added']
    genres = row['Type'].split(', ')

    # Check if any of the genres in this row are in our top 10 list
    for g in genres:
        if g in top_10_genre_names:
            # If it is, add it to our new list with its year
            genre_year_list.append({'Year_Added': year, 'Genre': g})
```

#### **Objective 3**: Compare country-wise contributions to Netflix’s catalog.

```python
# Use collections.Counter to count countries
country_counter = Counter()

# Loop through each row's 'Country' string
for countries in df['Country']:
    country_list = countries.split(', ')
    country_counter.update(country_list)

# *** Important: Remove the 'Unknown' category we added during cleaning ***
if 'Unknown' in country_counter:
    country_counter.pop('Unknown')
    print("Removed 'Unknown' category from country counts.")
```

```python
# Get the 10 most common countries
top_10_countries = country_counter.most_common(10)
# Convert to a DataFrame for easy plotting
df_top_10_countries = pd.DataFrame(top_10_countries, columns=['Country', 'Count'])

print("\n--- Top 10 Content Producing Countries on Netflix ---")
print(df_top_10_countries)
```

```python
# --- Visualization ---
plt.figure(figsize=(12, 8))

# Use seaborn for a horizontal bar plot
sns.barplot(x='Count', y='Country', data=df_top_10_countries, palette='viridis')

plt.title('Top 10 Content Producing Countries on Netflix', fontsize=16)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tight_layout()

# Save the plot
plt.savefig('top_10_countries.png')

print("\nVisualization saved as 'top_10_countries.png'")
```

**Outcome 3**- As Netflix is US company it is not surprising that it has most US shows. India, UK occupies 2nd and 3rd position respectively while Canada, France and Japan follows with almost 300+ Titles
